```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dfa=pd.read_csv("data2.csv")
dfa.isnull().sum()
```

```python

mode=dfa['WHEEZING'].mode()[0]
mode
```

```python
dfa['WHEEZING'] = dfa['WHEEZING'].fillna(value=mode)
```

```python
age_mode=dfa['AGE'].mode()[0]
dfa['AGE']=dfa['AGE'].fillna(value=age_mode)
```

```python
dfa['AGE'].min()
dfa['AGE'].max()
#age ranges from 21 to 87
```

```python
dfa.isnull().sum()
```

```python
sns.histplot(dfa['AGE'],kde=True,bins=25)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dfa['LUNG_CANCER'] = le.fit_transform(dfa['LUNG_CANCER'])
dfa['GENDER']= le.fit_transform(dfa['GENDER'])
dfa
```

```python
corr=dfa.corr()
corr
```

```python

plt.figure(figsize=(15,9))
sns.heatmap(corr,annot=True,cmap='mako')
```

```python
x=dfa.drop(columns='LUNG_CANCER')
y=dfa['LUNG_CANCER']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=200)
model.fit(x_train,y_train)
sc=model.score(x_test,y_test)
sc*100
#0.9354838709677419
```
